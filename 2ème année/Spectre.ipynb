```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm

angle='-45°'
a = np.load('amp_avg_7500test_code'+angle+'.npy')
p = np.load('phi_avg_7500test_code'+angle+'.npy')
c = []
for i in range(20):
    c.append(complex(a[i]*np.cos(p[i]),a[i]*np.sin(p[i])))   
a = np.concatenate((a,a))
p = np.concatenate((p,p))
c = np.concatenate((c,c))
T = np.arange(40)

#Plot de l'amplitude
plt.plot(T,a)
plt.title('Amplitude')
plt.xlabel('Temps')
plt.ylabel('Amplitude')
plt.savefig('Ampl'+angle)
plt.show()
plt.close()


#Plot de la phase        
plt.plot(T,p)
plt.title('Phase')
plt.xlabel('Temps')
plt.ylabel('Phase')
plt.savefig('Phase'+angle)
plt.show()
plt.close()

#Plot des spectres en fréquence
sp = np.fft.fft(c, n=40)
freq = np.fft.fftfreq(T.shape[-1])
sort = np.argsort([freq,sp])
vabs = np.abs(sp[sort[0]])
#vabs[19:22]=vabs[19:22]/10
plt.plot(freq[sort[0]], vabs)   
plt.title('enveloppe')
plt.xlabel('Freq')
plt.ylabel('abs(FFT)')
plt.savefig('Spectre'+angle)
plt.show()



```
